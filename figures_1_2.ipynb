```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#NB measure
def d1(x, y):
    num = 0.5*(abs(x - y) * (abs(x) + abs(y)))
    den = x*x + y*y
    return 0.0 if den == 0 else (num / den)

def Delta(x, y, z):
    return d1(x, z) - d1(x, y) - d1(y, z)
```

```python
#FIGURE 1
```

```python
# d(x,y) surface
L = 1000
grid = 250
xs = np.linspace(-L, L, grid)
ys = np.linspace(-L, L, grid)
X, Y = np.meshgrid(xs, ys)

D = np.vectorize(d1)(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
surf = ax.plot_surface(X, Y, D, linewidth=0, antialiased=True, cmap='plasma')

fig.colorbar(surf, ax=ax, shrink=0.65, aspect=10, pad=0.15, label="d(x,y)")
#fig.colorbar(surf, ax=ax, shrink=0.65, label="d(x,y)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("Dissimilarity d(x,y)")
ax.view_init(elev=20, azim=-45)
#plt.title("Dissimilarity between x and y under the proposed NB")


plt.savefig("figure1.png", dpi=300, bbox_inches="tight")
plt.savefig("figure1.tiff", dpi=300, bbox_inches="tight")
plt.savefig('figure1.eps', format='eps',bbox_inches="tight")


plt.show()
```

```python
#contuar
import numpy as np
import matplotlib.pyplot as plt
```

```python
#NB dissimilarity
def NB_dissimilarity(x, y):
    numerator = np.sum(np.abs(x - y) * (np.abs(x) + np.abs(y)))
    denominator = np.sum(x**2 + y**2)
    return (numerator / denominator)*(0.5) if denominator != 0 else 0
```

```python
# Create grid
x = np.linspace(-1, 1, 200)
y = np.linspace(-1, 1, 200)
X, Y = np.meshgrid(x, y)

# Compute dissimilarity
Z = np.zeros_like(X)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = NB_dissimilarity(X[i, j], Y[i, j])

# Plot contour with filled levels
plt.figure(figsize=(8, 6))
cp = plt.contourf(X, Y, Z, levels=50, cmap="plasma")
cbar = plt.colorbar(cp)
cbar.set_label("NB dissimilarity")
plt.title("Contour plot of NB dissimilarity between x and y")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)


plt.savefig("figure2.tiff", dpi=300)
plt.savefig('figure2.eps', format='eps')
plt.savefig('figure2.png', format='png')

plt.show()
```
